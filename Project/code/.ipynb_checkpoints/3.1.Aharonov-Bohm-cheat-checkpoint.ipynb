#### Definition of the graphene function with some parameters
* t is the energy
* a is the lattice parameter


```python
def make_system(a=1,R_ext=20,W=20, L=20,t=1.0):
    graphene = kwant.lattice.honeycomb(a,norbs=1)
    a, b = graphene.sublattices
    sys = kwant.Builder()
 
    #def rectangular(pos):
    #    x, y = pos
    #    return abs(x) < L//2 and abs(y) < W//2
    #def lead_shape(pos):
    #    x, y = pos
    #    return -W//2  < y < W//2
    def circle(pos): return (pos[0]**2 + pos[1]**2 <R_ext**2 and pos[0]**2 + pos[1]**2 >(R_ext-W)**2) #or (leadregion)
    #def inner_circle(pos):
    #    x, y=pos
    #return  x**2 + y**2 <R_in
        
    
    sys[graphene.shape(circle, (R_ext-W/2,0))] = 0
    #sys[graphene.shape(circle, (0,0))] = 0
    hoppings = (((0, 0), a, b), ((0, 1), a, b), ((-1, 1), a, b))
    sys[[kwant.builder.HoppingKind(*hopping) for hopping in hoppings]] = -t
    
    #del  sys[graphene.shape(inner_circle, (0,0))] 
    #sym0 = kwant.TranslationalSymmetry(graphene.vec((-1, 0)))
    #lead0 = kwant.Builder(sym0)
    #lead0[graphene.shape(lead_shape, (0, 0))] =  0
    #lead0[graphene.neighbors()]= -t
    #sys.attach_lead(lead0)
    
    #sym1 = kwant.TranslationalSymmetry(graphene.vec((1, 0)))
    #lead1 = kwant.Builder(sym1)
    #lead1[graphene.shape(lead_shape, (0, 0))] =  0
    #lead1[graphene.neighbors()]=-t
    #sys.attach_lead(lead1)

    return sys
```

```python
H=kwant.Builder()
a=1
R_ext=30
W=10
L=20
t=1.0
H=make_system(a,R_ext,W,L,t)
kwant.plot(H)
```
